# loading the dataset 

```python
import os

# Define the file path
file_path = "./Martial Peak [Complete] (F3).txt"

# Check if the file exists before attempting to read
if os.path.exists(file_path):
    try:
        # Open the file with utf-8 encoding
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()

        print(f"Successfully loaded {len(text)} characters.")

        # Print the first 500 characters to verify content
        print("\n--- First 500 Characters ---")
        print(text[:500])

    except Exception as e:
        print(f"An error occurred while reading the file: {e}")
else:
    print(f"File not found at: {file_path}")
```

# data length

```python
print("data length:", len(text))
```

# vocabulary size

```python
chars = sorted(list(set(text)))
vocab_size = len(chars)
print("all the unique characters:", ''.join(chars) )
print(" number of unique characters:", len(chars))
```

# tokenization 

```python
# string to character mapping
stoi = { ch:i for i,ch in enumerate(chars)}
# character to string mapping
itos = { i:ch for i,ch in enumerate(chars)}
encode = lambda s: [stoi[c] for c in s] # encoder: take a string, output a list of integers
decode = lambda l: ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string
```

# trying the tokenizer

```python
print(encode("yang kai"))
print(decode(encode("yang kai")))
```
